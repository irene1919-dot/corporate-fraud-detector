```python
# 1. Flag companies where Net Income is high, but they aren't generating real operational cash
data['Cash_Warning'] = (data['Net_Income'] > 0) & (data['Cash_Flow_Operating'] <= 0)

# 2. Flag companies with a dangerously high Debt-to-Equity ratio (above 2.5 is very risky)
data['Debt_Warning'] = data['Debt_to_Equity'] > 2.5

# 3. Create a master list of all suspicious companies flagged by our math rules
flagged_companies = data[(data['Cash_Warning'] == True) | (data['Debt_Warning'] == True)]

# 4. Print the total count of flagged companies
print(f"🕵️‍♂️ Your detective program flagged {len(flagged_companies)} companies out of {len(data)} total records!")
print("\n--- SNEAK PEEK OF SUSPICIOUS COMPANIES ---")

# 5. Display the key stats of the top 10 flagged companies
print(flagged_companies[['Net_Income', 'Cash_Flow_Operating', 'Debt_to_Equity', 'Financial_Status']].head(10))
```

```python
# 1. NEW STRICT RULE: Only flag if Net Income is positive, BUT Cash Flow is completely below zero (Negative)
data['Cash_Warning_Strict'] = (data['Net_Income'] > 0) & (data['Cash_Flow_Operating'] < 0)

# 2. NEW STRICT RULE: Raise the debt barrier. Only flag if Debt-to-Equity is over 4.0 (extremely dangerously high)
data['Debt_Warning_Strict'] = data['Debt_to_Equity'] > 4.0

# 3. CRITICAL COMBINATION: Only flag a company if they fail BOTH warnings at the exact same time
data['Master_Alert'] = (data['Cash_Warning_Strict'] == True) & (data['Debt_Warning_Strict'] == True)

# 4. Filter the data to see our new, highly-targeted suspects
strict_suspects = data[data['Master_Alert'] == True]

# 5. Print the new results
print(f"🕵️‍♂️ Tightened Net: Your program now only flagged {len(strict_suspects)} companies!")
print("\n--- NEW TARGETED SUSPECTS ---")
print(strict_suspects[['Net_Income', 'Cash_Flow_Operating', 'Debt_to_Equity', 'Financial_Status']].head(10))

```

```python
# 1. BALANCED CASH CHECK: Net income is high, but operational cash is deeply negative (less than -$100,000)
# (Assuming the numbers in the dataset are standard currency values)
data['Cash_Warning_Balanced'] = (data['Net_Income'] > 50000) & (data['Cash_Flow_Operating'] < -100000)

# 2. BALANCED DEBT CHECK: Raise the debt barrier to 3.5 (standard corporate danger zone)
data['Debt_Warning_Balanced'] = data['Debt_to_Equity'] > 3.5

# 3. COMBINE WITH "OR": Flag if they cross either of these higher, balanced barriers
data['Balanced_Alert'] = (data['Cash_Warning_Balanced'] == True) | (data['Debt_Warning_Balanced'] == True)

# 4. Filter the data to see our balanced suspects
balanced_suspects = data[data['Balanced_Alert'] == True]

# 5. Print the new results
print(f"🕵️‍♂️ Balanced Net: Your program flagged {len(balanced_suspects)} companies.")
print("\n--- BALANCED SUSPECTS PROFILE ---")
print(balanced_suspects[['Net_Income', 'Cash_Flow_Operating', 'Debt_to_Equity', 'Financial_Status']].head(15))

```

```python
# 1. Let the computer calculate the mathematical averages of the data
avg_debt = data['Debt_to_Equity'].mean()
avg_cash = data['Cash_Flow_Operating'].mean()

print(f"📊 Spreadsheet Averages -> Debt-to-Equity: {avg_debt:.2f} | Cash Flow: {avg_cash:.2f}")

# 2. ADVANCED DETECTIVE RULE:
# Flag companies where debt is 1.5 times worse than the average company,
# OR where cash flow is lower than the average company.
data['Smart_Debt_Warning'] = data['Debt_to_Equity'] > (avg_debt * 1.5)
data['Smart_Cash_Warning'] = data['Cash_Flow_Operating'] < avg_cash

# 3. Combine them using 'OR' so we catch outliers in either category
data['Smart_Alert'] = (data['Smart_Debt_Warning'] == True) | (data['Smart_Cash_Warning'] == True)

# 4. Filter the records
smart_suspects = data[data['Smart_Alert'] == True]

print(f"\n🕵️‍♂️ Smart Reset: Your program flagged {len(smart_suspects)} companies.")
print("\n--- NEW SMART SUSPECTS TARGETS ---")
print(smart_suspects[['Net_Income', 'Cash_Flow_Operating', 'Debt_to_Equity', 'Financial_Status']].head(10))
```

```python
# 1. Let's look at the unique labels inside the Financial_Status column
print("Here are the different types of company statuses in your data:")
print(data['Financial_Status'].value_counts())

# 2. Group the data by Financial_Status and calculate the average stats for each group
print("\n--- AVERAGE FINANCIAL STATS BY RISK GROUP ---")
profile = data.groupby('Financial_Status')[['Net_Income', 'Cash_Flow_Operating', 'Debt_to_Equity']].mean()
print(profile)
```

```python
# 1. Target the exact math profile of Anomaly companies (Debt_to_Equity below -0.01)
data['Final_Anomaly_Flag'] = data['Debt_to_Equity'] < -0.01

# 2. Target the exact math profile of High Risk companies (Net Income is higher than cash flow)
data['Final_Risk_Flag'] = data['Net_Income'] > data['Cash_Flow_Operating']

# 3. Create our final smart alert
data['Final_Alert'] = (data['Final_Anomaly_Flag'] == True) & (data['Final_Risk_Flag'] == True)

# 4. Filter the data to see our final group of suspects
final_suspects = data[data['Final_Alert'] == True]

# 5. Print the results to see our success rate!
print(f"🕵️‍♂️ Final Run: Your program isolated {len(final_suspects)} specific suspect companies.")
print("\n--- PERFORMANCE CHECK (What did your code actually catch?) ---")
print(final_suspects['Financial_Status'].value_counts())
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd

st.set_page_config(page_title="Fraud Detector", page_icon="🕵️‍♂️", layout="centered")
st.title("🕵️‍♂️ Corporate Fraud & Risk Detector")
st.write("Type a Company Row Number below to scan its financial report card for risks.")

# Load the data
try:
    data = pd.read_csv('Financial Statement Anomaly Dataset (1).csv')

    # Calculate our successful detective metrics from earlier
    data['Final_Anomaly_Flag'] = data['Debt_to_Equity'] < -0.01
    data['Final_Risk_Flag'] = data['Net_Income'] > data['Cash_Flow_Operating']
    data['Final_Alert'] = (data['Final_Anomaly_Flag'] == True) & (data['Final_Risk_Flag'] == True)

    # Input box for user
    company_idx = st.number_input("Enter Company Row Index (0 to 3999):", min_value=0, max_value=3999, value=0)

    if st.button("Run Financial Audit"):
        row = data.iloc[company_idx]
        st.subheader(f"📊 Audit Results for Company Row {company_idx}")

        # Display the core stats in clean boxes
        col1, col2, col3 = st.columns(3)
        col1.metric("Net Income", f"{row['Net_Income']:.4f}")
        col2.metric("Operating Cash", f"{row['Cash_Flow_Operating']:.4f}")
        col3.metric("Debt to Equity", f"{row['Debt_to_Equity']:.4f}")

        # Check if our detective flags triggered an alert
        if row['Final_Alert']:
            st.error("🚨 RED ALERT: High Risk Profile Detected! This company matches fraudulent metrics.")
        else:
            st.success("✅ GREEN LIGHT: This company looks stable based on current data patterns.")

        st.info(f"Actual Dataset Label for this Company: **{row['Financial_Status']}**")

except Exception as e:
    st.error(f"Could not load data file. Make sure it is uploaded! Error: {e}")
```
